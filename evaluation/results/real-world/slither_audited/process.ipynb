```python
import csv

# Define the input and output file names
input_file = './67/slither-report.csv'

# Specify the target filename (e.g., 'Maximillion.sol')
target_file = 'BaseStrategy'
target_filename = f'{target_file}.sol'

output_file = f'./67/{target_file}_output.csv'

with open(input_file, mode='r') as infile:
    reader = csv.DictReader(infile)
    
    # Open the output CSV file for writing
    with open(output_file, mode='w', newline='') as outfile:
        fieldnames = ['Severity', 'Description', 'Line']
        writer = csv.DictWriter(outfile, fieldnames=fieldnames)
        
        # Write the header to the output file
        writer.writeheader()
        
        # Iterate over each row in the input file
        for row in reader:
            # Extract and trim the 'Filename' column
            filename = row['Filename'].strip()
            
            # Debug: Print the filename to check its content
            print(f"Checking filename: {filename}")
            
            # Check if the target substring is in the 'Filename' column
            if target_filename in filename:
                # Debug: Print a message when a match is found
                print(f"Match found for '{target_filename}' in row: {row}")
                
                # Write the relevant information to the output file
                writer.writerow({
                    'Severity': row['Severity'],
                    'Description': row['Description'],
                    'Line': row['Line']
                })

print(f"Filtered descriptions for '{target_filename}' have been extracted to {output_file}")
```
